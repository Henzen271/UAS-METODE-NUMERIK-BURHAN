```python
# Fungsi f(x)
def f(x):
    # Parameter berdasarkan NIM
    k = (22 % 10) + 1  # k = 3
    c = (22 // 10) + 1 # c = 3
    return x**3 - k*x + c

# Metode Bisection
def bisection_method(a, b, tol):
    if f(a) * f(b) >= 0:
        return "Interval tidak valid. f(a) dan f(b) harus memiliki tanda berbeda."

    while abs(b - a) > tol:
        mid = (a + b) / 2
        if f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid
    return (a + b) / 2

# Interval awal dan toleransi galat
a, b = -3, -2
tol = 1e-5

# Menjalankan metode bisection
root = bisection_method(a, b, tol)
print(f"Akar fungsi adalah: {root}")
```
